Import STOPWORDS
to clean the word cloud from common or irrelevant words

```python
from wordcloud import STOPWORDS
```

```python
stopwords = set(STOPWORDS)
```

my Custom Stopwords
# Add custom words to exclude from the word cloud

```python
stopwords.update([
    "Aboderin", "Oyekunle", "Bankole", "OS", "FA", "Olakojo", "Olaoye",
    "Saminu", "Submitted", "Journal", "Agriculture", "Science",
    "Environment", "Thailand", "Peruvian", "Proceedings",
    "2021", "2023", "2024", "2025"
])
```

# To confirm if the word being filtered is still in the stopwords

```python
print("Aboderin" in stopwords) 
```

```python
print("Maize" in stopwords)
```

#### Generate a Simple Word Cloud


```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud(
    stopwords=stopwords  # remove the words we defined
).generate(text)
```

#### Display the word cloud

```python
plt.imshow(wordcloud, interpolation='bilinear')  # smooth image
plt.axis("off")  # remove axes
plt.show()
```

#### Create a Customized Word Cloud

```python
wordcloud = WordCloud(
    stopwords=stopwords,
    max_font_size=70,        # Controls largest word size
    background_color="white", # Clean white background
    width=800,               # Image width (increase for better resolution)
    height=500,              # Image height
    max_words=100            # Maximum number of words displayed
).generate(text)
```

##### Diplay the new wordcloud

```python
plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("wordcloud_olawale.png")
```

##### Word Cloud Using a Shape

```python
from PIL import Image
import numpy as np
```

#### Load image and convert to numpy array


```python
mask = np.array(Image.open("Turtle.png"))
```

```python
plt.imshow(mask)
plt.axis("off")
plt.show()
```

#### Generate word cloud with shape

```python
wc = WordCloud(
    background_color="white",
    max_words=1000,
    mask=mask,
    stopwords=stopwords,
    contour_width=2,
    contour_color="grey",
    repeat=True,
    max_font_size=50
)
```

```python
wc.generate(text)
```

```python
wc.to_file("wordcloud_shaped.png")
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

### Q1: The world cloud you just produced is based on how many variables?
- The word cloud is based on two variables: 
1. The words (terms) extracted from the text
2. The frequency of each word (how many times each word appears)

### Q2: Which is the type of each variable in your word cloud?
1. Words  : Categorical variable (qualitative). Each word represents a category
2. Frequency: Numerical variable (quantitative)  because it counts how many times each word appears
